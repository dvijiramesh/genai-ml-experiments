# Zero-Shot Prompting with Large Language Models

This notebook demonstrates zero-shot prompting, where a large language model is given a task without prior examples and is expected to infer the correct behavior directly from the instruction.

The examples illustrate how LLMs can perform tasks such as classification and text generation using only task instructions.

```python
import os
from openai import OpenAI

client = OpenAI()
```

```python
def get_completion(prompt, model="gpt-3.5-turbo"):
    try:
        messages = [{"role": "user", "content": prompt}]

        response = client.chat.completions.create(
            model=model,
            messages=messages,
            temperature=0.1,
            top_p=0.8,
            max_tokens=512
        )

        return response

    except Exception as e:
        return {"error": f"Error: {str(e)}"}
```


- Define a simple prompt for the model

```python
prompt = "Explain artificial intelligence"
response = get_completion(prompt)

if hasattr(response, "choices"):
    print(response.choices[0].message.content)
else:
    print("Error:", response)
```

- You can further enhance the function to include structured conversations

```python
def get_completion(prompt, model="gpt-3.5-turbo"):

    messages = [
        {"role": "system", "content": "You are a math tutor"},
        {"role": "user", "content": "Explain me the ..."},
        {"role": "assistant", "content": "The concept is like..."},
        {"role": "user", "content": "Give me an example"},
    ]

    response = client.chat.completions.create(
        model=model,
        messages=messages,
        temperature=0.9,
        top_p=0.8,
        max_tokens=512
    )

    return response.choices[0].message.content
```

- Call the function to generate the response

```python
prompt = "explain trignometry"
response = get_completion(prompt)
print(response)
```

## Observations

Zero-shot prompting works well for simple tasks without requiring examples.
However, the output format may vary depending on how clearly the prompt specifies the expected response.
